```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
file1=pd.read_csv('Bihar.csv')
```

```python
file1
```

```python
df=pd.DataFrame(file1)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df.Date,df.Confirmed,color="pink",label='Confirmed Cases')
plt.plot(df.Date,df.Cured,color="navy",label='Cured Cases')
plt.plot(df.Date,df.Death, color="orange",label='Death Cases')
plt.legend()
plt.bar(df.Date,df.Confirmed,label='Confirmed Cases',color='lightskyblue')
plt.bar(df.Date,df.Cured,color="pink",label='Cured Cases')
plt.bar(df.Date,df.Death,color="red",label='Death Cases')
plt.title("COVID-19 DAILY ANALYSIS:BIHAR")
plt.xticks(rotation=90)
plt.xlabel=("Date")
plt.ylabel=("Number of Cases")
plt.savefig("major_daily.png")
plt.show()
```

```python
file2=pd.read_csv('Weekly Bihar.csv')
```

```python
df2=pd.DataFrame(file2)
```

```python
plt.figure(figsize=(20,10))
style.use('ggplot')
x1=df2.Date
y1=df2.Confirmed
x2=df2.Date
y2=df2.Death
x3=df2.Date
y3=df2.Cured
plt.plot(x1,y1,'red',label='Confirmed', linestyle='-',marker='o')
plt.plot(x2,y2,'blue',label='Deaths', linestyle='-',marker='o')
plt.plot(x3,y3,'green',label='Recovered', linestyle='-',marker='o')
plt.xlabel=("Cases")
plt.ylabel=("Days")
plt.title("WEEKLY ANALYSIS OF BIHAR")
plt.legend()
plt.xticks(rotation=90)
plt.show
plt.savefig("Weekly_major.png")
```
