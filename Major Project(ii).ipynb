```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
dataset1 = pd.read_csv('Bihar.csv') 
```

```python
dataset1
```

```python
heat_map = sns.heatmap(dataset1.corr(), annot=True)
plt.title("HEATMAP-Bihar")
plt.savefig("heatmap_bihar1_major.png")
plt.show()
```

```python
d = pd.pivot_table(dataset1,index=['State'],
                    values='Confirmed',
                    columns='Date' )
sns.heatmap(d)
plt.title("HEATMAP-Bihar(Confirmed Cases)")
plt.savefig("heatmap_Bihar_major.png")
```

```python
dataset2=pd.read_csv('state.csv')
```

```python
heat_map = sns.heatmap(dataset2.corr(), annot=True)
plt.title("HEATMAP-India")
plt.savefig("heatmap_india_major.png")
plt.show()
```
